```python
def myScore(Target, Pred):

    tmp1 = 0

    tmp2 = 0

    for i in range(len(Target)):

        if Target[i] > Pred[i]:

            tmp1 = tmp1 + math.exp((-Pred[i] + Target[i]) / 13.0) - 1

        else:

            tmp2 = tmp2 + math.exp((Pred[i] - Target[i]) / 10.0) - 1

    tmp = tmp1 + tmp2

    return tmp


def rmse(y_true, y_pred): # 评估函数rmse

    return K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1))


def custom_loss_function(y_true, y_pred, alpha = 0.7):

    y_error = y_pred - y_true 

    # mse损失 

    loss_mse = K.mean(K.square(y_error), axis=-1) 

    # score损失 

    bool_idx = K.greater(y_error, 0) 

    score_loss1 = K.exp(-1 * y_error / 13) - 1 

    score_loss2 = K.exp(y_error / 10) - 1

    loss_score = K.sum(K.switch(bool_idx, score_loss2, score_loss1)) 

    # 加权损失 

    return alpha * loss_score + (1 - alpha) * loss_mse # 返回 "标量" 自定义损失值 
```

```python
# 自身索引打乱两次，对生成的两个数据进行mixup 
def mixup(x, y):  

    # 打乱索引列表
    def shuffle_index_fun(index): 

        random.shuffle(index) 

        return index.copy()

    rd = 0.5

    shuffle_index = list(range(x.shape[0]))

    r_index1 = shuffle_index_fun(shuffle_index)

    r_index2 = shuffle_index_fun(shuffle_index)

    x_mixup = (1-rd) * x[r_index1] + rd * x[r_index2]

    y_mixup = (1-rd) * y[r_index1] + rd * y[r_index2]

    x_new, y_new = np.vstack((x, x_mixup)), np.vstack((y, y_mixup))

    return x_new, y_new  
```

### 全连接网络

```python
Image(filename = '../picture/vae.png')
```

```python
## Encoder  

inputs = Input(shape=(winSize, Feasize), name='encoder_input')

x = Dense(hidden_dim, activation='relu')(inputs) 

z_mean = Dense(latent_dim, name='z_mean')(x)

z_log_var = Dense(latent_dim, name='z_log_sigma')(x)

## Sampler 

def sampling(args):

    z_mean, z_log_var = args

    epsilon = K.random_normal(shape=(winSize, latent_dim))

    return z_mean + K.exp(z_log_var / 2) * epsilon

z = Lambda(sampling, name='z')([z_mean, z_log_var]) 

encoder = Model(inputs, [z_mean, z], name='encoder')

## Decoder 

latent_inputs = Input(shape=(winSize, latent_dim), name='z_sampling') 

x = Dense(hidden_dim, activation='relu')(latent_inputs)

x_hat = Dense(Feasize, activation='sigmoid', name='decoder_output')(x)

decoder = Model(latent_inputs, x_hat, name='decoder')
 
# VAE model 

outputs = decoder(encoder(inputs)[1]) 

vae = Model(inputs, outputs, name='vae_mlp')

# loss

reconstruction_loss = mse(inputs, outputs) 

kl_loss = - 0.5 * K.mean(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var))

vae_loss = reconstruction_loss + kl_loss

vae.add_loss(vae_loss)

vae.compile(optimizer='Adam') #  metrics=[rmse] 

pt = ReduceLROnPlateau(monitor='loss', verbose=0, mode='min', factor=0.1, patience=30) 

er = EarlyStopping(monitor='loss', verbose=0, mode='min', patience=50)

cbks = [pt, er]  

vae.fit(trainX_vae, trainX_vae, epochs=500, batch_size=128, verbose=1, callbacks=cbks, shuffle=True) 
```

### LSTM

```python
## Encoder  

inputs = Input(shape=(winSize, Feasize), name='encoder_input')

x = LSTM(hidden_dim, return_sequences=True)(inputs) 

z_mean_lstm = Dense(latent_dim)(x)

z_log_var_lstm = Dense(latent_dim)(x)

## Sampler 

def sampling(args):

    z_mean_lstm, z_log_var_lstm = args

    epsilon = K.random_normal(shape=(winSize, latent_dim))

    return z_mean_lstm + K.exp(z_log_var_lstm / 2) * epsilon

z = Lambda(sampling, name='z')([z_mean_lstm, z_log_var_lstm]) 

encoder_lstm = Model(inputs, [z_mean_lstm, z], name='encoder_lstm')

## Decoder 

latent_inputs = Input(shape=(winSize, latent_dim), name='z_sampling') # z

x = LSTM(hidden_dim, return_sequences=True)(latent_inputs)

x_hat = LSTM(Feasize, return_sequences=True, name='decoder_output')(x)

decoder_lstm = Model(latent_inputs, x_hat, name='decoder_lstm')
 
# VAE model 

outputs = decoder_lstm(encoder_lstm(inputs)[1]) 

vae_lstm = Model(inputs, outputs, name='vae_lstm')

# print(vae_lstm.summary())

## loss 

reconstruction_loss = mse(inputs, outputs) 

kl_loss = - 0.5 * K.mean(1 + z_log_var_lstm - K.square(z_mean_lstm) - K.exp(z_log_var_lstm))

vae_loss = reconstruction_loss + kl_loss

vae_lstm.add_loss(vae_loss)

vae_lstm.compile(optimizer='adam') 

pt = ReduceLROnPlateau(monitor='loss', verbose=0, mode='min', factor=0.1, patience=30) 

er = EarlyStopping(patience=50, verbose=0, mode='min', monitor='loss')

cbks = [pt, er] 

history = vae_lstm.fit(trainX_vae, trainX_vae, epochs=500, batch_size=128, verbose=1, callbacks=cbks)
```

```python
Mish = lambda x:x * K.tanh(K.softplus(x)) 

def attention_block(inputs, dim): # (None, 32, 10)

    a_probs = Dense(dim, activation='softmax')(inputs) 
    
    output_attention_mul = multiply([inputs, a_probs])  
    
    return output_attention_mul 

def conv_block(inputs, filters, dilation = None): 
    
    if dilation == None: # 特征维度
        
        c1 = TimeDistributed(Conv1D(filters, kernel_size=5, strides=1, padding='same'))(inputs) 

        n1 = BatchNormalization(momentum=0.6)(c1) 

        a1 = Activation(activation=Mish)(n1) 

        c2 = TimeDistributed(Conv1D(filters, kernel_size=5, strides=1, padding='same'))(a1) 

        n2 = BatchNormalization(momentum=0.6)(c2)

        a2 = Activation(activation=Mish)(n2)
        
    else: # 时间维度
        
        c1 = TimeDistributed(Conv1D(filters, kernel_size=5, strides=1, padding='causal', dilation_rate=dilation))(inputs) # 扩张的因果卷积

        n1 = BatchNormalization(momentum=0.6)(c1)

        a1 = Activation(activation=Mish)(n1)

        c2 = TimeDistributed(Conv1D(filters, kernel_size=5, strides=1, padding='causal', dilation_rate=dilation))(a1) 

        n2 = BatchNormalization(momentum=0.6)(c2)

        a2 = Activation(activation=Mish)(n2)

    return a2 

def residual_block(inputs, filters, factor = None): 
    
    if factor == None: # 特征维度

        n1 = conv_block(inputs, filters) 

        residual = TimeDistributed(SeparableConv1D(1, kernel_size=1, padding='same'))(n1) # （None, 32, 10, 64） -->  （None, 32, 10, 1）

        outputs = add([inputs, residual]) 
    
    else: # 时间维度
        
        dilation = 2 ** factor

        n1 = conv_block(inputs, filters, dilation) 

        residual = TimeDistributed(SeparableConv1D(1, kernel_size=1, padding='same'))(n1) 

        outputs = add([inputs, residual]) 

    return outputs

def TCN_Net(inputs, is_fea = True): # (None, 32, 10)
    
    if is_fea == True:

        TIME_STEPS = int(inputs.shape[1])

        Feasize = int(inputs.shape[2]) 

        x = attention_block(inputs, Feasize) # (None, 32, 10) 输出与输入维度相同

        x = Reshape((TIME_STEPS, Feasize, 1))(x) # (None, 32, 10, 1)  

        x0 = residual_block(x, 64) 

        x1 = residual_block(x0, 64) 

        x1 = Reshape((TIME_STEPS, Feasize))(x1) # (?, 32, 10) 

        x1 = attention_block(x1, Feasize) 
        
    else:
    
        Feasize = int(inputs.shape[1]) 

        TIME_STEPS = int(inputs.shape[2])

        x = attention_block(inputs, TIME_STEPS) # (None, 14, 32) 

        x = Reshape((Feasize, TIME_STEPS, 1))(x) # (None, 14, 32, 1)

        x0 = residual_block(x, 64, 0) 

        x1 = residual_block(x0, 64, 1) 

        x1 = Reshape((Feasize, TIME_STEPS))(x1) # (?, 14, 32) 

        x1 = attention_block(x1, TIME_STEPS)
    
    return x1

def FC_block(input):

    X = Dense(100, activation='relu')(input) # (None, 100) 

    X = Dropout(0.2)(X)

    X = Dense(20, activation='relu')(X) # (None, 20) 

    X = Dropout(0.2)(X) 

    output = Dense(1, activation='linear')(X) # (None, 1)

    return output

def PA_TCN_Net():

    inputs = Input(shape=(winSize, Feasize, 1)) # (None, 32, 14, 1)

    x1 = Reshape((winSize, Feasize))(inputs) # (None, 32, 14)  

    x1 = encoder(x1)[0] 

    x2 = Permute((2, 1))(x1) # (None, 14, 32) 

    fea_out = TCN_Net(x1, True) # (None, 32, 14)

    time_out = TCN_Net(x2, False) 

    time_flatten = Flatten()(time_out) # (None, 448)

    fea_flatten = Flatten()(fea_out) # (None, 448)

    X = Concatenate(axis=-1)([time_flatten, fea_flatten]) # flatten_tcn, flatten_tcn_reverse --> gap_tcn, gap_tcn_reverse

    output = FC_block(X) # (None, 1)

    return Model(inputs, output) 
```

## 模型训练

```python
max_RUL = 125.0

RMSE_t = []

score_t = []

ttime = []

for i in range(10):  # run 10 times   

    print('iteration: ', i)

    start1 = time.time()

    model = PA_TCN_Net() 

    # print(model.summary())

    # graph_path = model_path + 'TCN_model_time.png' 

    # plot_model(model, to_file = graph_path, show_shapes = True)  # 绘制模型图

    model.compile(optimizer='adam', loss=custom_loss_function, metrics=[rmse]) 

    pt = ReduceLROnPlateau(monitor='val_loss', verbose=0, mode='min', factor=0.1, patience=8) 

    er = EarlyStopping(patience=12, verbose=0, restore_best_weights=True, mode='min', monitor='val_loss') 

    cp = ModelCheckpoint(model_path+'Net{}.h5'.format(str(i)), monitor='val_loss', mode='min', save_best_only=True) # 保存所有epoch model中best model

    cbks = [pt, er, cp] 

    ##########################用来确定最好的mixup数据##############################

    trainX_new_mixup, trainY_mixup = mixup(trainX_new, trainY) 

    np.save(precessData_path + 'trainX_new_mixup{}_FD00{}.npy'.format(str(i), str(FD_id)), trainX_new_mixup)

    np.save(precessData_path + 'trainY_mixup{}_FD00{}.npy'.format(str(i), str(FD_id)), trainY_mixup)

    # trainX_new_mixup = np.load(precessData_path + 'trainX_new_mixup{}_FD00{}.npy'.format(str(2), str(FD_id)))

    # trainY_mixup = np.load(precessData_path + 'trainY_mixup{}_FD00{}.npy'.format(str(2), str(FD_id)))

    ###############################################################################

    model.fit(trainX_new_mixup, 
              trainY_mixup, 
              batch_size=128, 
              validation_split=0.1, 
              verbose=1, 
              epochs=200, 
              shuffle=True, 
              callbacks=cbks)  ## 原数据 trainX_new, trainY  mixup数据 trainX_new_mixup, trainY_mixup 

    model.load_weights(model_path+'Net{}.h5'.format(str(i)))

    yPreds = model.predict(testX_new) # (100， 1) # testX / testX_new

    del model

    gc.collect()

    end1 = time.time()

    run_time = (end1 - start1) / 60
    print('该循环程序运行时间：', run_time, '分钟')     

    ttime.append(run_time)

    yPreds = yPreds.ravel() # (100, ) 

    yPreds = max_RUL * yPreds # 因为标签是进行归一化之后的，所以预测得到的结果要乘以max_RUL

    test_rmse = np.sqrt(mean_squared_error(max_RUL * testY, yPreds))

    test_score = myScore(max_RUL * testY, yPreds)

    print('lastScore:', test_score, 'lastRMSE', test_rmse)

    RMSE_t.append(test_rmse)

    score_t.append(test_score)

rmse_mean = np.mean(RMSE_t)

rmse_std = np.std(RMSE_t)

score_mean = np.mean(score_t)

score_std = np.std(score_t)

ttime_mean = np.mean(ttime)

print('rmse_mean:', rmse_mean, 'rmse_std:', rmse_std, 'score_mean:', score_mean, 'score_std:', score_std)

print('ttime_mean:', ttime_mean)
```
